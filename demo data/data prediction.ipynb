```python
import pandas as pd
import numpy as np

# Set the random seed for reproducibility
np.random.seed(41)

# Define the number of records
num_records = 2000

# Generate random data for the dataset
loan_ids = ['L{:03d}'.format(i) for i in range(1, num_records + 1)]
genders = np.random.choice(['Male', 'Female'], num_records)
married_status = np.random.choice(['Yes', 'No'], num_records)
dependents = np.random.choice(['0', '1', '2', '3+'], num_records)
education = np.random.choice(['Graduate', 'Not Graduate'], num_records)
self_employed = np.random.choice(['Yes', 'No'], num_records)
applicant_income = np.random.randint(10000, 500000, num_records)
coapplicant_income = np.random.randint(0, 500000, num_records)
loan_amount = np.random.randint(500, 5000000, num_records)
loan_term = np.random.choice(range(1, 5), num_records)
credit_history = np.random.randint(0, 951, num_records)
property_area = np.random.choice(['Urban', 'Semiurban', 'Rural'], num_records)

# Generate "Loan_Status" based on the specified criteria
rate_of_interest = 6 / 12 / 100
months = loan_term * 12
rate1 = ((1 + rate_of_interest) ** months) / ((1 + rate_of_interest) ** months - 1)
emi = loan_amount * rate_of_interest * rate1

loan_status = np.where(np.logical_or(credit_history > 650, ((applicant_income + coapplicant_income) / 12) > emi), 'Y', 'N')

# Create a DataFrame from the generated data
loan_dataset = pd.DataFrame({
    'Loan_ID': loan_ids,
    'Gender': genders,
    'Married': married_status,
    'Dependents': dependents,
    'Education': education,
    'Self_Employed': self_employed,
    'ApplicantIncome': applicant_income,
    'CoapplicantIncome': coapplicant_income,
    'LoanAmount': loan_amount,
    'Loan_Amount_Term': loan_term,
    'Credit_History': credit_history,
    'Property_Area': property_area,
    'Loan_Status': loan_status
})

# Save the DataFrame to a CSV file
loan_dataset.to_csv('demo.csv', index=False)
```
